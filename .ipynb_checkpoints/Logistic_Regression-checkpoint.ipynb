```python
import pandas as pd
import numpy as np

from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score

# --------------------------------------------------
# 1. Load data
# --------------------------------------------------
df = pd.read_csv("Tommy_Award_Player_Game_Table.csv")

# Keep only players who actually played
df = df[df["minutes_decimal"] > 0].copy()

# --------------------------------------------------
# 2. Engineer simple features
# --------------------------------------------------
mins = df["minutes_decimal"].replace(0, np.nan)

df["stocks"] = df["steals"] + df["blocks"]
df["points_per_min"] = df["points"] / mins
df["oreb_per_min"] = df["reboundsOffensive"] / mins
df["reb_per_min"] = df["reboundsTotal"] / mins
df["ast_per_min"] = df["assists"] / mins
df["stocks_per_min"] = df["stocks"] / mins
df["hustle_proxy"] = df["reboundsOffensive"] + df["steals"] + df["blocks"]

# --------------------------------------------------
# 3. Add within-game ranks
# --------------------------------------------------
rank_cols = [
    "points",
    "reboundsOffensive",
    "reboundsTotal",
    "assists",
    "steals",
    "blocks",
    "plusMinusPoints",
    "minutes_decimal",
    "stocks",
    "hustle_proxy"
]

for col in rank_cols:
    df[f"{col}_rank"] = df.groupby("GAME_ID")[col].rank(
        ascending=False,
        method="min"
    )

# --------------------------------------------------
# 4. Choose features
# --------------------------------------------------
feature_cols = [
    "minutes_decimal",
    "points",
    "reboundsOffensive",
    "reboundsDefensive",
    "reboundsTotal",
    "assists",
    "steals",
    "blocks",
    "turnovers",
    "foulsPersonal",
    "plusMinusPoints",
    "fieldGoalsMade",
    "fieldGoalsAttempted",
    "threePointersMade",
    "threePointersAttempted",
    "freeThrowsMade",
    "stocks",
    "points_per_min",
    "oreb_per_min",
    "reb_per_min",
    "ast_per_min",
    "stocks_per_min",
    "hustle_proxy",
    "points_rank",
    "reboundsOffensive_rank",
    "reboundsTotal_rank",
    "assists_rank",
    "steals_rank",
    "blocks_rank",
    "plusMinusPoints_rank",
    "minutes_decimal_rank",
    "stocks_rank",
    "hustle_proxy_rank"
]

X = df[feature_cols].copy()
y = df["y"].copy()
groups = df["GAME_ID"].copy()

# --------------------------------------------------
# 5. Split by game, not by row
# --------------------------------------------------
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_train = X.iloc[train_idx]
X_test = X.iloc[test_idx]
y_train = y.iloc[train_idx]
y_test = y.iloc[test_idx]

train_df = df.iloc[train_idx].copy()
test_df = df.iloc[test_idx].copy()

print("Train games:", train_df["GAME_ID"].nunique())
print("Test games:", test_df["GAME_ID"].nunique())

# --------------------------------------------------
# 6. Build logistic regression pipeline
# --------------------------------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, feature_cols)
])

model = Pipeline(steps=[
    ("prep", preprocessor),
    ("clf", LogisticRegression(max_iter=3000))
])

# --------------------------------------------------
# 7. Train
# --------------------------------------------------
model.fit(X_train, y_train)

# --------------------------------------------------
# 8. Predict probabilities
# --------------------------------------------------
test_df["pred_prob"] = model.predict_proba(X_test)[:, 1]

# --------------------------------------------------
# 9. Evaluate properly by game
# --------------------------------------------------

# Top-1 winner prediction:
pred_winners = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(1)
           .copy()
)

game_level_accuracy = pred_winners["y"].mean()

# Top-3 accuracy:
top3 = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(3)
           .copy()
)

top3_accuracy = top3.groupby("GAME_ID")["y"].max().mean()

# Row-level metrics (secondary, not main)
row_logloss = log_loss(y_test, test_df["pred_prob"])
row_auc = roc_auc_score(y_test, test_df["pred_prob"])

print("\n--- Results ---")
print("Game-level winner accuracy:", round(game_level_accuracy, 4))
print("Top-3 accuracy:", round(top3_accuracy, 4))
print("Row-level log loss:", round(row_logloss, 4))
print("Row-level AUC:", round(row_auc, 4))

# --------------------------------------------------
# 10. Show a few sample predicted games
# --------------------------------------------------
sample_games = test_df["GAME_ID"].drop_duplicates().head(5).tolist()

for gid in sample_games:
    game_rows = (
        test_df[test_df["GAME_ID"] == gid]
        .sort_values("pred_prob", ascending=False)
        [["GAME_ID", "player_name", "y", "pred_prob", "points", "reboundsOffensive",
          "reboundsTotal", "assists", "steals", "blocks", "plusMinusPoints"]]
    )
    print(f"\nGAME_ID: {gid}")
    print(game_rows.head(5).to_string(index=False))
```
