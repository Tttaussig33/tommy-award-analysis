# Pregame Logistic Regression

This notebook trains and evaluates a logistic regression model using `Tommy_Award_Pregame_Features.csv`.

It uses a chronological train/test split by game date so the evaluation stays realistic for a pregame prediction setup.

```python
import pandas as pd
from IPython.display import display
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

INPUT_PATH = "Tommy_Award_Pregame_Features.csv"
TEST_SIZE = 0.20


def load_dataset(path: str = INPUT_PATH) -> pd.DataFrame:
    df = pd.read_csv(path, dtype={"GAME_ID": str}).copy()
    df["game_date"] = pd.to_datetime(df["game_date"])
    return df


def get_feature_columns(df: pd.DataFrame) -> list[str]:
    excluded_cols = {
        "GAME_ID",
        "game_date",
        "season",
        "personId",
        "player_name",
        "player_name_key",
        "winner_names",
        "y",
    }
    return [col for col in df.columns if col not in excluded_cols]


def chronological_game_split(
    df: pd.DataFrame,
    test_size: float = TEST_SIZE,
) -> tuple[pd.DataFrame, pd.DataFrame]:
    game_order = (
        df[["GAME_ID", "game_date"]]
        .drop_duplicates()
        .sort_values(["game_date", "GAME_ID"], kind="mergesort")
        .reset_index(drop=True)
    )

    split_idx = int(len(game_order) * (1 - test_size))
    split_idx = max(1, min(split_idx, len(game_order) - 1))

    train_game_ids = set(game_order.iloc[:split_idx]["GAME_ID"])
    test_game_ids = set(game_order.iloc[split_idx:]["GAME_ID"])

    train_df = df[df["GAME_ID"].isin(train_game_ids)].copy()
    test_df = df[df["GAME_ID"].isin(test_game_ids)].copy()

    return train_df, test_df
```

```python
df = load_dataset()
feature_cols = get_feature_columns(df)

for col in feature_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

train_df, test_df = chronological_game_split(df)

X_train = train_df[feature_cols]
X_test = test_df[feature_cols]
y_train = train_df["y"]
y_test = test_df["y"]

print(f"Rows: {len(df)}")
print(f"Features: {len(feature_cols)}")
print(f"Train games: {train_df['GAME_ID'].nunique()}")
print(f"Test games: {test_df['GAME_ID'].nunique()}")

feature_cols
```

```python
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ]
)

model = Pipeline(
    steps=[
        ("prep", ColumnTransformer([("num", numeric_transformer, feature_cols)])),
        ("clf", LogisticRegression(max_iter=3000)),
    ]
)

model.fit(X_train, y_train)
test_df["pred_prob"] = model.predict_proba(X_test)[:, 1]

test_df[["GAME_ID", "game_date", "player_name", "winner_names", "y", "pred_prob"]].head()
```

```python
pred_winners = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
    .groupby("GAME_ID")
    .head(1)
    .copy()
)

top3 = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
    .groupby("GAME_ID")
    .head(3)
    .copy()
)

game_level_accuracy = pred_winners["y"].mean()
top3_accuracy = top3.groupby("GAME_ID")["y"].max().mean()
row_logloss = log_loss(y_test, test_df["pred_prob"])
row_auc = roc_auc_score(y_test, test_df["pred_prob"])

results_df = pd.DataFrame(
    {
        "metric": [
            "game_level_winner_accuracy",
            "top_3_accuracy",
            "row_level_log_loss",
            "row_level_auc",
        ],
        "value": [
            game_level_accuracy,
            top3_accuracy,
            row_logloss,
            row_auc,
        ],
    }
)

results_df["value"] = results_df["value"].round(4)
display(results_df)
```

```python
show_cols = [
    "GAME_ID",
    "game_date",
    "player_name",
    "winner_names",
    "y",
    "pred_prob",
]

sample_games = test_df["GAME_ID"].drop_duplicates().head(5).tolist()

for game_id in sample_games:
    game_rows = (
        test_df[test_df["GAME_ID"] == game_id]
        .sort_values("pred_prob", ascending=False)[show_cols]
        .head(5)
    )
    print(f"\nGAME_ID: {game_id}")
    display(game_rows)
```
