```python
import pandas as pd
import numpy as np

from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score

# --------------------------------------------------
# 1. Load data
# --------------------------------------------------
df = pd.read_csv("Tommy_Award_Player_Game_Table.csv")

# Keep only players who actually played
df = df[df["minutes_decimal"] > 0].copy()

# --------------------------------------------------
# 2. Engineer simple features
# --------------------------------------------------
mins = df["minutes_decimal"].replace(0, np.nan)

df["stocks"] = df["steals"] + df["blocks"]
df["points_per_min"] = df["points"] / mins
df["oreb_per_min"] = df["reboundsOffensive"] / mins
df["reb_per_min"] = df["reboundsTotal"] / mins
df["ast_per_min"] = df["assists"] / mins
df["stocks_per_min"] = df["stocks"] / mins
df["hustle_proxy"] = df["reboundsOffensive"] + df["steals"] + df["blocks"]

# --------------------------------------------------
# 3. Add within-game ranks
# --------------------------------------------------
rank_cols = [
    "points",
    "reboundsOffensive",
    "reboundsTotal",
    "assists",
    "steals",
    "blocks",
    "plusMinusPoints",
    "minutes_decimal",
    "stocks",
    "hustle_proxy"
]

for col in rank_cols:
    df[f"{col}_rank"] = df.groupby("GAME_ID")[col].rank(
        ascending=False,
        method="min"
    )

# --------------------------------------------------
# 4. Choose features
# --------------------------------------------------
feature_cols = [
    "minutes_decimal",
    "points",
    "reboundsOffensive",
    "reboundsDefensive",
    "reboundsTotal",
    "assists",
    "steals",
    "blocks",
    "turnovers",
    "foulsPersonal",
    "plusMinusPoints",
    "fieldGoalsMade",
    "fieldGoalsAttempted",
    "threePointersMade",
    "threePointersAttempted",
    "freeThrowsMade",
    "stocks",
    "points_per_min",
    "oreb_per_min",
    "reb_per_min",
    "ast_per_min",
    "stocks_per_min",
    "hustle_proxy",
    "points_rank",
    "reboundsOffensive_rank",
    "reboundsTotal_rank",
    "assists_rank",
    "steals_rank",
    "blocks_rank",
    "plusMinusPoints_rank",
    "minutes_decimal_rank",
    "stocks_rank",
    "hustle_proxy_rank"
]

X = df[feature_cols].copy()
y = df["y"].copy()
groups = df["GAME_ID"].copy()

# --------------------------------------------------
# 5. Split by game, not by row
# --------------------------------------------------
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_train = X.iloc[train_idx]
X_test = X.iloc[test_idx]
y_train = y.iloc[train_idx]
y_test = y.iloc[test_idx]

train_df = df.iloc[train_idx].copy()
test_df = df.iloc[test_idx].copy()

print("Train games:", train_df["GAME_ID"].nunique())
print("Test games:", test_df["GAME_ID"].nunique())

# --------------------------------------------------
# 6. Build logistic regression pipeline
# --------------------------------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, feature_cols)
])

model = Pipeline(steps=[
    ("prep", preprocessor),
    ("clf", LogisticRegression(max_iter=3000))
])

# --------------------------------------------------
# 7. Train
# --------------------------------------------------
model.fit(X_train, y_train)

# --------------------------------------------------
# 8. Predict probabilities
# --------------------------------------------------
test_df["pred_prob"] = model.predict_proba(X_test)[:, 1]

# --------------------------------------------------
# 9. Evaluate properly by game
# --------------------------------------------------

# Top-1 winner prediction:
pred_winners = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(1)
           .copy()
)

game_level_accuracy = pred_winners["y"].mean()

# Top-3 accuracy:
top3 = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(3)
           .copy()
)

top3_accuracy = top3.groupby("GAME_ID")["y"].max().mean()

# Row-level metrics (secondary, not main)
row_logloss = log_loss(y_test, test_df["pred_prob"])
row_auc = roc_auc_score(y_test, test_df["pred_prob"])

print("\n--- Results ---")
print("Game-level winner accuracy:", round(game_level_accuracy, 4))
print("Top-3 accuracy:", round(top3_accuracy, 4))
print("Row-level log loss:", round(row_logloss, 4))
print("Row-level AUC:", round(row_auc, 4))

# --------------------------------------------------
# 10. Show a few sample predicted games
# --------------------------------------------------
sample_games = test_df["GAME_ID"].drop_duplicates().head(5).tolist()

for gid in sample_games:
    game_rows = (
        test_df[test_df["GAME_ID"] == gid]
        .sort_values("pred_prob", ascending=False)
        [["GAME_ID", "player_name", "y", "pred_prob", "points", "reboundsOffensive",
          "reboundsTotal", "assists", "steals", "blocks", "plusMinusPoints"]]
    )
    print(f"\nGAME_ID: {gid}")
    print(game_rows.head(5).to_string(index=False))

row_level_accuracy = ((test_df["pred_prob"] >= 0.5).astype(int) == y_test).mean()
print("\nRow-level accuracy:", round(row_level_accuracy, 4))
```

```python
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score

# -----------------------------
# 1. Load your pregame dataset
# -----------------------------
df = pd.read_csv("Tommy_Award_Pregame_Features.csv")

# Make sure date is a real datetime
df["game_date"] = pd.to_datetime(df["game_date"])

# -----------------------------
# 2. Pick the feature columns
# -----------------------------
feature_cols = [
    "games_played_before",

    "season_avg_points",
    "points_last_3",
    "points_last_5",
    "points_last_10",

    "season_avg_offensive_rebounds",
    "offensive_rebounds_last_3",
    "offensive_rebounds_last_5",
    "offensive_rebounds_last_10",

    "season_avg_rebounds",
    "rebounds_last_3",
    "rebounds_last_5",
    "rebounds_last_10",

    "season_avg_assists",
    "assists_last_3",
    "assists_last_5",
    "assists_last_10",

    "season_avg_steals",
    "steals_last_3",
    "steals_last_5",
    "steals_last_10",

    "season_avg_blocks",
    "blocks_last_3",
    "blocks_last_5",
    "blocks_last_10",

    "season_avg_minutes",
    "minutes_last_3",
    "minutes_last_5",
    "minutes_last_10",

    "season_avg_plus_minus",
    "plus_minus_last_3",
    "plus_minus_last_5",
    "plus_minus_last_10",

    "season_avg_three_pm",
    "three_pm_last_3",
    "three_pm_last_5",
    "three_pm_last_10",

    "season_avg_stocks",
    "stocks_last_3",
    "stocks_last_5",
    "stocks_last_10",

    "season_avg_hustle_proxy",
    "hustle_proxy_last_3",
    "hustle_proxy_last_5",
    "hustle_proxy_last_10",
]

X = df[feature_cols].copy()
y = df["y"].copy()

# -----------------------------
# 3. Time-based split
# Train on older seasons, test on newest season
# -----------------------------
train_df = df[df["season"] != "2025-26"].copy()
test_df  = df[df["season"] == "2025-26"].copy()

X_train = train_df[feature_cols].copy()
y_train = train_df["y"].copy()

X_test = test_df[feature_cols].copy()
y_test = test_df["y"].copy()

print("Train games:", train_df["GAME_ID"].nunique())
print("Test games:", test_df["GAME_ID"].nunique())
print("Train rows:", len(train_df))
print("Test rows:", len(test_df))

# -----------------------------
# 4. Build logistic regression pipeline
# -----------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value=0)),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, feature_cols)
])

model = Pipeline(steps=[
    ("prep", preprocessor),
    ("clf", LogisticRegression(max_iter=5000))
])

# -----------------------------
# 5. Train
# -----------------------------
model.fit(X_train, y_train)

# -----------------------------
# 6. Predict probabilities
# -----------------------------
test_df = test_df.copy()
test_df["pred_prob"] = model.predict_proba(X_test)[:, 1]

# -----------------------------
# 7. Evaluate by game
# -----------------------------
pred_winners = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(1)
           .copy()
)

game_level_accuracy = pred_winners["y"].mean()

top3 = (
    test_df.sort_values(["GAME_ID", "pred_prob"], ascending=[True, False])
           .groupby("GAME_ID")
           .head(3)
           .copy()
)

top3_accuracy = top3.groupby("GAME_ID")["y"].max().mean()

row_logloss = log_loss(y_test, test_df["pred_prob"])
row_auc = roc_auc_score(y_test, test_df["pred_prob"])

print("\n--- Logistic Regression Results ---")
print("Game-level winner accuracy:", round(game_level_accuracy, 4))
print("Top-3 accuracy:", round(top3_accuracy, 4))
print("Row-level log loss:", round(row_logloss, 4))
print("Row-level AUC:", round(row_auc, 4))

# -----------------------------
# 8. Look at a few example games
# -----------------------------
sample_games = test_df["GAME_ID"].drop_duplicates().head(5).tolist()

for gid in sample_games:
    game_rows = (
        test_df[test_df["GAME_ID"] == gid]
        .sort_values("pred_prob", ascending=False)
        [["GAME_ID", "game_date", "player_name", "y", "pred_prob",
          "points_last_5", "stocks_last_5", "minutes_last_5", "hustle_proxy_last_5"]]
    )
    print(f"\nGAME_ID: {gid}")
    print(game_rows.head(5).to_string(index=False))
```
