# March 6, 2026 game — sample table (DataFrame)

Loads `Tommy_Award_Player_Game_Table_hustle.csv` from this project directory, keeps rows with **`game_date` = 2026-03-06** (March 6, 2026), and **displays** a small summary table the same way as the rest of your notebooks (interactive HTML in Jupyter).

**Tommy winner** is the `y` column shown as **1** (awarded the Tommy that game) or **0** (not). **Date** is repeated on every row.

The CSV has no `net_rating` column; **Net rtg.** uses `plusMinusPoints`, the same proxy as `net_rating` in the feature notebooks when that column is imputed.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display

ROOT = Path(".").resolve()
CSV_PATH = ROOT / "Tommy_Award_Player_Game_Table_hustle.csv"

df = pd.read_csv(CSV_PATH)
if "game_date" not in df.columns:
    raise ValueError("Expected column game_date in CSV.")

df["game_date"] = pd.to_datetime(df["game_date"], errors="coerce", format="mixed")
TARGET = pd.Timestamp("2026-03-06")
sub = df.loc[df["game_date"].dt.normalize() == TARGET.normalize()].copy()
if sub.empty:
    raise ValueError(
        f"No rows for game_date == {TARGET.date()} in {CSV_PATH.name}. "
        "Add that game to the CSV, or change TARGET in this cell."
    )
if "minutes_decimal" in sub.columns and sub["minutes_decimal"].notna().any():
    sub = sub.sort_values("minutes_decimal", ascending=False, na_position="last")

date_str = TARGET.strftime("%Y-%m-%d")
player_col = "player_name" if "player_name" in sub.columns else "nameI"
if player_col not in sub.columns:
    raise ValueError("Need player_name or nameI in CSV.")
if "y" not in sub.columns:
    raise ValueError("Expected column 'y' (Tommy winner / model target) in CSV.")

sub["net_rating"] = pd.to_numeric(
    sub["net_rating" if "net_rating" in sub.columns else "plusMinusPoints"],
    errors="coerce",
)

tommy_winner = sub["y"].fillna(0).astype(int)

show = pd.DataFrame(
    {
        "Date": [date_str] * len(sub),
        "Player": sub[player_col].astype(str),
        "Tommy winner": tommy_winner,
        "Min": sub["minutes_decimal"].round(1) if "minutes_decimal" in sub else float("nan"),
        "PTS": sub["points"] if "points" in sub else float("nan"),
        "REB": sub["reboundsTotal"] if "reboundsTotal" in sub else float("nan"),
        "AST": sub["assists"] if "assists" in sub else float("nan"),
        "OREB": sub["reboundsOffensive"] if "reboundsOffensive" in sub else float("nan"),
        "STL": sub["steals"] if "steals" in sub else float("nan"),
        "BLK": sub["blocks"] if "blocks" in sub else float("nan"),
        "Net rtg.": sub["net_rating"].round(0).astype("Int64"),
    }
)

gid = int(sub["gameId"].iloc[0]) if "gameId" in sub else None
print(f"Sample game: Boston Celtics box-style rows ({date_str}, game {gid})")
display(show)
```
